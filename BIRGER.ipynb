# BIRGER Project Proposal

This project involves the development of dynamic dashboards, real-time Analytics and AI insights through smart KPIs for BIRGER. support and service center (NOC).

Our Network Operations Centre monitors the after-sales service of BIRGER. for all its customer in the region. NOC responds to our customer service and support request calls, allocates incidents to our field engineers, ensures that all incidents are addressed promptly and reports back to our customers.

Our qualified and experienced engineers provide 1st and 2nd line hardware maintenance services, software maintenance, field and remote diagnostics and problem resolution, and network monitoring.

BIRGER. already has an incident reporting platform for the logging and follow-up of incidents reported by our customers. The platform is two tier architecture with a frontend web base application and a backend SQL database.

The main objective of this project is to leverage the existing data to design dynamic dashboards, real-time Analytics and AI insights through smart KPIs for our support and service centre (NOC) to enhance our efficiency, decision-making processes, and overall performance. Hence assisting us in providing a better proximity service for a better customer experience.

# Scatter Plot of Number of Incidents by Month and Geo with their Model No

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you already loaded your dataset into a DataFrame called df

# Convert 'Created On' to 'Month' and format it to show only the month abbreviation
df['Month'] = pd.to_datetime(df['Created On']).dt.strftime('%b')

# Define the custom order of months
custom_month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Set the 'Month' column data type and order
df['Month'] = pd.Categorical(df['Month'], categories=custom_month_order, ordered=True)

# Group the data by 'Month' and 'Geo' and count the number of incidents
geo_month_counts = df.groupby(['Month', 'Geo']).size().unstack().fillna(0)

# Create a color map for the different Geo values
colors = {'Mauritius': 'red', 'Seychelles': 'blue', 'Rodrigues': 'green'}

# Plot the scatter plot for each Geo
for geo, color in colors.items():
    plt.scatter(geo_month_counts.index, geo_month_counts[geo], label=geo, c=color, alpha=0.6, s=50)

    # Add 'Model No' annotations on top of the dots (only for available data points)
    for i, txt in enumerate(df[df['Geo'] == geo]['Model No']):
        if i < len(geo_month_counts):
            plt.annotate(txt, (geo_month_counts.index[i], geo_month_counts[geo].iloc[i]), ha='center')

# Set labels and title
plt.xlabel('Month')
plt.ylabel('Number of Incidents')
plt.title('Number of Incidents by Month and Geo')
plt.legend(title='Geo', loc='upper right')

# Show the plot
plt.show()
```
